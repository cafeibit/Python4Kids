## Data Gathering and Analysis Sample: Toronto Weather Data Analysis
### 1. How to gather data online to save as cvs file
> From <a href="https://www.ncdc.noaa.gov/cdo-web/">National Centers for Environmental Information</a> you can download cities/countries weather infomation with CSV format. From <a href="https://climate.weather.gc.ca">Canada Historical Climate Data</a> you can download any cities in Canada.

### 2. How to gather data from file
> We’ll use Python’s csv module to process weather data stored in the CSV (comma-separated values) format and analyze high and low temperatures over time in two different locations. We’ll then use Matplotlib to generate a chart based on our downloaded data to display variations in temperature in two dissimilar environments: Sitka, Alaska, and Death Valley, California. Then we use the json module to access earthquake data stored in the JSON format and use Plotly to draw a world map showing the locations and magnitudes of recent earthquakes. 

<b>Plotting Dates</b>
> Now we can improve our temperature data plot by extracting dates (the datetime Module) for the daily highs and passing those highs and dates to plot(). 

<b>Plotting Data in a Temperature Chart</b>
> To visualize the temperature data we have, we’ll first create a simple plot of the daily highs using Matplotlib.

```python
import csv
from datetime import datetime

from matplotlib import pyplot as plt

filename = 'data/NewYorkWeatherData2010.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)

    # Get dates, and high and low temperatures from this file.
    dates, highs, lows = [], [], []
    for row in reader:
        current_date = datetime.strptime(row[1], '%Y%m%d')
        try:
            high = row[4]
            low = row[5]
        except ValueError:
            print(f"Missing data for {current_date}")
        else:
            dates.append(current_date)
            highs.append(high)
            lows.append(low)

# Plot the high and low temperatures.
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.plot(dates, highs, c='red', alpha=0.5)
ax.plot(dates, lows, c='blue', alpha=0.5)
plt.fill_between(dates, highs, lows, facecolor='blue', alpha=0.1)

# Format plot.
title = "Daily high and low temperatures - 2010\nNew York, NY"
plt.title(title, fontsize=20)
plt.xlabel('', fontsize=16)
fig.autofmt_xdate()
plt.ylabel("Temperature (F)", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)

plt.show()
```
